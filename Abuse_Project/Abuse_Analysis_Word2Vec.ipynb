```python
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
abuse_reple=pd.read_csv('./Abuse_reple.csv')
del abuse_reple['Unnamed: 0']

abuse_reple_list=list(abuse_reple['x'])
word = [a.split(' ') for a in abuse_reple_list]

print(word)
```

```python
from konlpy.tag import Twitter

Twitter = Twitter()
# 파일을 다시 처음부터 읽음.
#Twitter.pos(abuse_word, norm=True, stem=True) # 단어 토큰화
result=[]
for i in abuse_reple_list:
    a=Twitter.pos(i, norm=True, stem=True)
    result.append(a)
result
```
